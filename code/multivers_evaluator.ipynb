# Load data

```python
actual_values = []
predicted_values = []
arr1 = []
arr2 = []
with open(f'/content/gdrive/MyDrive/Independent study - Max & Carlos/data/results/multivers_test.jsonl') as f_pdf: ###### WILL NEED TO CHANGE NAMES
    for line in f_pdf:
        diction = json.loads(line)
        arr1.append(diction)
with open(f'/content/gdrive/MyDrive/Independent study - Max & Carlos/data/results/multivers_results.jsonl') as f_pdf: ###### WILL NEED TO CHANGE NAMES
    for line in f_pdf:
        diction = json.loads(line)
        arr2.append(diction)
```

```python
print(len(arr1), len(arr2))
```

# Evaluate

```python
actual_values = []
predicted_values = []

for i, item in enumerate(arr1): # SUPP, CONTRA, UNREL
    for doc_id in item['doc_ids']:
        if str(doc_id) in item['evidence'] and item['evidence'][str(doc_id)][0]['label'] == "SUPPORT":
            actual_values.append(0)
        elif str(doc_id) in item['evidence'] and item['evidence'][str(doc_id)][0]['label'] == "CONTRADICT":
            actual_values.append(1)
        else:
            actual_values.append(2)

        if str(doc_id) in arr2[i]['evidence'] and arr2[i]['evidence'][str(doc_id)]['label'] == "SUPPORT":
            predicted_values.append(0)
        elif str(doc_id) in arr2[i]['evidence'] and arr2[i]['evidence'][str(doc_id)]['label'] == "CONTRADICT":
            predicted_values.append(1)
        else:
            predicted_values.append(2)
```

```python
print(len(actual_values))
print(len(predicted_values))
```

```python
print(f1_score(actual_values, predicted_values, average='micro'),
      f1_score(actual_values, predicted_values, average='macro'))
```
